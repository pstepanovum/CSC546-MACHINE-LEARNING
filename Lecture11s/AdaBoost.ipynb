```python
%matplotlib inline
from IPython import display
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
sn.set(style='whitegrid')
from sklearn import datasets, svm
from matplotlib.patches import Ellipse
```

```python
from sklearn.datasets import make_moons
Xmoon, ymoon = make_moons(200, noise=.05, random_state=0)
plt.scatter(Xmoon[ymoon==0, 0], Xmoon[ymoon==0, 1], color='r', marker = '^');
plt.scatter(Xmoon[ymoon==1, 0], Xmoon[ymoon==1, 1], color='b', marker = 's');
plt.title('two classes (red triangles and blue squras) are NOT linearly separable', fontsize=16)
plt.axis([-2, 3, -1, 1.5])
```

```python
from sklearn.ensemble import AdaBoostClassifier
ada=AdaBoostClassifier()
ada.fit(Xmoon,ymoon)
```

```python
xx, yy = np.meshgrid(np.arange(-2, 3, 0.01), np.arange(-1, 1.5, 0.01))
xx_r=xx.reshape(xx.shape[0]*xx.shape[1],1)
yy_r=yy.reshape(yy.shape[0]*yy.shape[1],1)
xx_yy = np.concatenate([xx_r, yy_r], axis=1)
Z = ada.predict(xx_yy)
Z = Z.reshape(xx.shape)
# Put the result into a color plot
plt.pcolormesh(xx, yy, Z,cmap='viridis')
plt.scatter(Xmoon[ymoon ==0, 0], Xmoon[ymoon ==0, 1], color='r', marker = '^');
plt.scatter(Xmoon[ymoon ==1, 0], Xmoon[ymoon ==1, 1], color='b', marker = 's');
```
