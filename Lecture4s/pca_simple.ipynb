```python
from sklearn.decomposition import PCA
import numpy as np
```

```python
X=np.random.rand(20, 8)
X
```

```python
model=PCA(n_components=3, whiten=False) # output will be beta
```

```python
model.fit(X)
```

```python
beta=model.transform(X)
```

```python
beta.shape
```

```python
lamda=model.explained_variance_
lamda
```

```python
y=beta/np.sqrt(lamda)
y.shape
```
